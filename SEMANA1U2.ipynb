###CONCEPTOS BASICOS DE CONJUNTOS
En la teoria de conjuntos, trabajamos con colecciones de elementos que comparten caracteristicas en común, estos pueden ser números, letras, objetos, resultados de experimentos, etc. En el contetexto de los juegos de rolde mesa como AoS Soulbound, estos elemntos pueden ser los resultados posibles al realizar una tirada de dados de 6 caras

---
notación: LOs conjuntos se denotan con letras mayusculas(por ejemplo A) Y SUS elementos se escriben entre llaves, separados por comas ejemplo: A={1,2,3,4,5,6}
En donde A representa el conjunto de resultados posibles al lanzar un dado de 6 caras
Ejemplo en soulbound
Si lanzas un dado para determinar el exito de un ataque, el conjunto de resultados posibles es: R= {1,2,3,4,5,6}
Si un ataque tiene un Dn 5 , el resultados de exitos es E={5,6}

#universo
El universo es el conjunto que contiene todos los elementos en un contexto dado, se denota como U
Ejemplo: en el lanzamiento anterior los resultados representan el universo posible U={1,2,3,4,5,6}
El universo serian todos los resultados posibles. EL axioma es el resultado.

#conjunto vacio
Es conjunto que no tiene elementos. Se denota como 0 ó {}
En souldbound, si un personaje tiene una habilidad que nunca falla, el conjunnto de fallos seria: Fallos=0.
#subconjunto
Un subconjunto es un conjunto que contiene algunos elementos del universo. Se denota como A pertenece a U .Un conjunto de A es subconjunto de un conjunto B si todos los elementos de A estan en B
Ejemplo: si definos el conjunto de A como "numeros pares", entonces
A={2,4,6} pertenece a U={1,2,3,4,5,6}
Aqui A es un subconjunto de U porque todos sus elementos estan en U
Si definimos un ataque con DN 4 siendo B={4,5,6} siendo este el conjunto de exitos posibles, entonces podemos determinaar que el conjunto que el conjunto definido como "el mayor posible del dado"
C={6} pertenece a B={4,5,6}
donde C es el conjunto de criticos posibles y pertenece en su totalidad a B
#interseccion de conjuntos
La interseccion de dos conjuntos A y B= {5,6} (exitos en una posible DN=5)
Entonces: A n B = {6}
En Souldbound, si un personaje necesita un resultado par y mayor a 4 para activar una habilidad la interseccion seria:
EXITOS={5,6} n {2,4,6} = {6}
#UNION DE CONJUNTOS
La union de dos conjuntos A y B es el conjunto de elementos que pertenece al menos unos de los conjuntos. Se denota como A u B
EJEMPLO= Si A={2,4,6} (resultados pares) y B= {5,6}(exitos en un ataque), entonces A u B ={2,4,5,6}
#diferencia de conjuntos
La diferencia entre dos conjuntos A y B es el conjunto de elementos que estan en A pero no en B se denota como A-B
#complemento de un conjunto
El complemento de un conjunto A es el conjunto de elementos que no estan en A, pero si en el universo U. Se denota como A o Ac..
